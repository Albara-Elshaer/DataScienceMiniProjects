## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott', 'age': 500},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]
        },
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich', 'age': 5},
         'counties': [{'name': 'Dade', 'population': 40000},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, ['counties'])
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname',['info', 'governor'], ['info', 'age']])
```

## 2. Finding the top 10 major project themes (using column 'mjtheme_namecode'). The code below is a visiulaization using a bar plot to show major theme (or "sector") the projects are involved in

## 3. Above you will notice that some entries have only the code and the name is missing. Below is where I create a dataframe with the missing names filled in.

```python
# Creating a new dataframe to get an overall glimpse of which codes have the missing names 
df3=json_normalize(wbp, 'mjtheme_namecode')
df3
```

```python
#iterate through each row of dataframe        
df4= df3.drop_duplicates(subset=None, keep='first', inplace=False).query('name != "" ')
```

```python
# Using the merge function to fill in the missing values 
df5=df4.merge(df3, how='right', on='code')

df5['name_x'].value_counts()
```

```python
# Creating a visiualization using a plot bar 
df5['name_x'].value_counts().plot.barh()
```
