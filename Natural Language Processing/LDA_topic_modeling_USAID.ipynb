```python
# Step 2: Clean the corpus (input: usaid_corpus.txt)(output: clean_usaid_corpus)

# import spaCy 
import spacy 

nlp = spacy.load("en_core_web_sm")


# TODO: read the usaid_corpus into a variable 

# TODO: load corpus into spaCy (output: tokenized_usaid_corpus) 

# open the usaid_corpus
tokenized_usaid_corpus = nlp(mexico)
    


# import stopword library 
from nltk.corpus import stopwords

# create stop word set
stop_words = set(stopwords.words('english'))

# use the stopword list to filter out stopwords in tokenized_usaid_corpus (output: clean_tokenized_usaid_corpus) 
# Filter out punctuations
clean_tokenizied_usaid_corpus = []

for token in tokenized_usaid_corpus:
    if token.is_alpha and not token.is_punct and token.text not in stop_words:
        clean_tokenizied_usaid_corpus.append(token.lemma_)  
```

```python
# Step 3: Prepare BOW (input: clean_corpus) (output: BOW)

# import gensim library
import gensim
from gensim import corpora

# pass the cleaned_tokenized_usaid_corpus.txt to the gensim library (output: gensim object)
usaid_unique_token_dictionary  = corpora.Dictionary([clean_tokenizied_usaid_corpus])

# create the bow from the genism library 

# save the BOW (either a varible or a file)
usaid_bow = usaid_unique_token_dictionary.doc2bow(clean_tokenizied_usaid_corpus)

# print BOW
usaid_bow
```

```python
# Step 4: Run the LDA model (input: BOW) (output: list of topics for each document)

# import the LDA model 
from gensim.models.ldamulticore import LdaMulticore

from gensim.models import CoherenceModel
from gensim.models import LdaModel

# Enable logging for gensim - optional
#import logging
#logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.ERROR)

# pass bow to the model 
# lda = LdaMulticore(usaid_bow,num_topics=10, id2word=usaid_unique_token_dictionary, workers=3, passes=1)

# Set training parameters.
num_topics = 10
chunksize = 2000
passes = 1
iterations = 400
eval_every = None  # Don't evaluate model perplexity, takes too much time.


model = LdaModel(
    corpus=usaid_bow,
    id2word=usaid_unique_token_dictionary,
    chunksize=chunksize,
    alpha='auto',
    eta='auto',
    iterations=iterations,
    num_topics=num_topics,
    passes=passes,
    eval_every=eval_every
)
# save the output 
```
